```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os

df = pd.read_csv("/investments_VC.csv", encoding="latin1")
```

```python
#show first few rows
df.head()
```

```python
# Check for missing values in each column
print("Missing Values Per Column:\n", df.isnull().sum())
# Check column data types
print("\nColumn Data Types:\n", df.dtypes)
# Check for duplicate rows
print("\nNumber of Duplicate Rows:", df.duplicated().sum())
# Display first 5 rows
df.head()
```

```python
print(df.columns.tolist())
```

```python
# Remove leading and trailing spaces from all column names
df.columns = df.columns.str.strip()

# Print the updated column names to verify the change
print(df.columns.tolist())
```

```python
# Check for missing values in each column again
print(df.isnull().sum())
```

```python
# Drop rows where critical columns are missing (name, category, funding amount)
df = df.dropna(subset=["name", "category_list", "funding_total_usd", "market"])

# Check missing values again
print(df.isnull().sum())
```

```python
# Fill missing categorical values with "Unknown"
df["status"] = df["status"].fillna("Unknown")
df["country_code"] = df["country_code"].fillna("Unknown")
df["city"] = df["city"].fillna("Unknown")

# Check missing values again
print(df.isnull().sum())
```

```python
# Fill missing homepage URLs with "No Website"
df["homepage_url"] = df["homepage_url"].fillna("No Website")

# Check missing values again
print(df.isnull().sum())
```

```python
df["founded_at"] = df["founded_at"].fillna("Unknown")

# Check missing values again
print(df.isnull().sum())
```

```python
# Find unique values that are not numbers
print(df["funding_total_usd"].unique()[:20])  # Show the first 20 unique values
```

```python
# Replace non-numeric values with NaN
df["funding_total_usd"] = df["funding_total_usd"].replace([" - ", "", "unknown"], np.nan)

# Remove commas and safely convert to float
df["funding_total_usd"] = df["funding_total_usd"].str.replace(",", "", regex=True)

# Convert column to float
df["funding_total_usd"] = pd.to_numeric(df["funding_total_usd"], errors='coerce')

# Confirm the conversion
print(df["funding_total_usd"].dtype)
```

Explarotary Data Analysis (EDA)

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Group by Industry and sum funding amounts
industry_funding = df.groupby("market")["funding_total_usd"].sum().sort_values(ascending=False).head(10)

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(x=industry_funding.index, y=industry_funding.values, palette="coolwarm")
plt.xticks(rotation=45, ha='right')
plt.xlabel("Industry")
plt.ylabel("Total Investment Amount (USD)")
plt.title("Top 10 Funded Industries")
plt.show()
```

This graph presents the distribution of total investment across different industries, highlighting which sectors draw the most venture capital. Biotechnology stands out as the top recipient, reflecting its high development costs and the potential for groundbreaking advancements in medical research and pharmaceuticals. The Mobile and Software industries follow, fueled by the rise of smartphones, cloud computing, and AI-driven innovations. HealthTech and Clean Technology also secure significant funding, emphasizing the growing focus on healthcare improvements and sustainable solutions. The fairly balanced allocation among leading industries suggests that venture capital is spreading across various technology-driven sectors rather than being concentrated in just one

```python
# Group by Country and sum funding amounts
country_funding = df.groupby("country_code")["funding_total_usd"].sum().sort_values(ascending=False).head(10)

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(x=country_funding.index, y=country_funding.values, palette="coolwarm")
plt.xticks(rotation=45, ha='right')
plt.xlabel("Country")
plt.ylabel("Total Investment Amount (USD)")
plt.title("Top 10 Countries by Startup Funding")
plt.show()
```

This visualization showcases the global distribution of startup funding, with the United States leading by a wide margin. Its dominance is driven by Silicon Valley, a well-established venture capital ecosystem, and a strong culture of innovation. China ranks second, largely due to government-backed tech investments and the rapid expansion of AI and FinTech. The UK, India, and Canada also attract substantial funding, solidifying their positions as emerging tech hubs with robust startup ecosystems. Notably, the inclusion of an "Unknown" category suggests possible data inconsistencies or startups operating across multiple regions without a clear country designation

```python
# Convert 'first_funding_at' to datetime format
df["first_funding_at"] = pd.to_datetime(df["first_funding_at"], errors="coerce")

# Extract the year from 'first_funding_at'
df["funding_year"] = df["first_funding_at"].dt.year

# Group by year and sum funding amounts
yearly_funding = df.groupby("funding_year")["funding_total_usd"].sum()

# Plot
plt.figure(figsize=(12, 6))
sns.lineplot(x=yearly_funding.index, y=yearly_funding.values, marker="o", color="b")
plt.xlabel("Year")
plt.ylabel("Total Investment Amount (USD)")
plt.title("Startup Funding Over Time")
plt.grid(True)
plt.show()
```

The funding trend over time shows a gradual rise until the late 1990s, followed by a sharp increase. The early 2000s surge aligns with the dot-com boom, while the subsequent decline reflects the dot-com bust. A notable funding spike occurs after 2005, driven by the emergence of social media, cloud computing, and AI-focused companies. The peak around 2010 suggests a period of heavy investment, likely fueled by mobile technology growth, advancements in deep learning, and the rise of unicorn startups like Uber and Airbnb. Post-2015 fluctuations indicate evolving investment patterns, economic cycles, and shifts in venture capital priorities

```python
# Group by industry and year to see trends
industry_yearly_funding = df.groupby(["funding_year", "market"])["funding_total_usd"].sum().reset_index()

# Filter for the top 5 industries
top_industries = df.groupby("market")["funding_total_usd"].sum().sort_values(ascending=False).head(5).index
industry_yearly_funding = industry_yearly_funding[industry_yearly_funding["market"].isin(top_industries)]

# Plot
plt.figure(figsize=(12, 6))
sns.lineplot(data=industry_yearly_funding, x="funding_year", y="funding_total_usd", hue="market", marker="o")
plt.xlabel("Year")
plt.ylabel("Total Investment Amount (USD)")
plt.title("Funding Trends Over Time for Top 5 Industries")
plt.legend(title="Industry")
plt.grid(True)
plt.show()
```

This graph offers insight into the evolution of funding trends across major industries. Software and Mobile experienced rapid growth in the 2000s, with Mobile reaching a peak around 2010, likely driven by the smartphone boom and app-based business models. Biotechnology and HealthTech saw steady increases, reflecting continued investment in medical innovations, personalized healthcare, and biotech advancements. Clean Technology, while growing, appears more volatile, influenced by government policies, sustainability trends, and shifts in renewable energy investments. The varied funding patterns suggest that investment waves in different industries are shaped by technological advancements, market demands, and regulatory factors
